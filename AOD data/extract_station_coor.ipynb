```python
from meteostat import Stations
import pandas as pd
```

```python

# 1. Initialize the Stations class
stations = Stations()

# 2. Filter stations by Vietnam's country code (VN)
# The .region() method accepts country and an optional state/region code.
vietnam_stations = stations.region('VN')

# 3. Fetch the data as a Pandas DataFrame
# This DataFrame contains all the metadata for the stations.
station_data = vietnam_stations.fetch()

# 4. Filter the DataFrame to only keep the station ID and coordinates
# The relevant columns are 'id', 'latitude', and 'longitude'.
coordinates = station_data[['name', 'latitude', 'longitude', 'elevation']]

# 5. Print the result
column_names = station_data.columns.tolist()
print(column_names)
print(coordinates)

```

```python
# Optional: Save the data to a CSV file
coordinates.to_csv('vietnam_meteostat_stations.csv', index=True)
```
